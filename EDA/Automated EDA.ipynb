```python
import ydata_profiling #pandas_profiling
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
from pandas_profiling import ProfileReport #import ProfileReport library from ydata profiling
```

```python
profile=ProfileReport(df,explorative=True,dark_mode=True) #save the data in a variable
```

```python
#convert to HTML
profile.to_file('Auromated_EDA_output.html')
```
