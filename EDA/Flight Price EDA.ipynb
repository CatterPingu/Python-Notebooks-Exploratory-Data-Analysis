```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_df=pd.read_excel('Flight_Price_Train.xlsx')
train_df.head()
```

```python
test_df=pd.read_excel('Flight_Price_Test.xlsx')
test_df.head()
```

```python
#Combining both the datasets
final_df=pd.concat([train_df,test_df],axis=0)

final_df.head()
```

```python
final_df.info()
```

```python
final_df['Date_of_Journey'].str.split('/')

#We'll choose the indexes as date month and year i.e. 0,1 and 2
```

```python
#Feature Engineering

#Creating three features from date of journey as day month and year

final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
#Converting the newly created columns to integer

final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
final_df.head()

#dropping the original date of journey column as it is no longer requeired
```

```python
#Now we will handle arrival time

#Here we need to remove the date from it as we already have the date
```

```python
final_df['Arrival_Time'].str.split(' ')
```

```python
final_df['Arrival_Time'].str.split(' ')[0].str[0]
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]

final_df.head()
```

```python
final_df.isnull().sum()
```

```python
#Now we will split the arrival time further into hours and minutes

final_df['Hours']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Minutes']=final_df['Arrival_Time'].str.split(':').str[1]
final_df.head()
```

```python
#Now we convert both these colummns to integer type

final_df['Hours']=final_df['Hours'].astype(int)
final_df['Minutes']=final_df['Minutes'].astype(int)
```

```python
final_df.info()
```

```python
#Droppinf the arrival time feature as it isn't required

final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df.rename(columns={"Hours": "Arrival_Hours", "Minutes": "Arrival_Minutes"},inplace=True)
```

```python
final_df.head()
```

```python
#Now doing the process for departure time

final_df['Dep_Hours']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_Minutes']=final_df['Dep_Time'].str.split(':').str[1]
print(final_df.head())

final_df['Dep_Hours']=final_df['Dep_Hours'].astype(int)
final_df['Dep_Minutes']=final_df['Dep_Minutes'].astype(int)
print(final_df.info())
```

```python
final_df.head()
```

```python
#Dropping the departure time column

final_df.drop('Dep_Time',axis=1,inplace=True)
final_df.head()
```

```python
#Handling total_stops, converting everything into int/float

final_df[final_df['Total_Stops'].isnull()]

#Only this column has null value, since delhi to cochin is taking 23hours, we may replace it by 2 or 3 stops
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'2 stops':2,'1 stop':1,'3 stops':3,'nan':2,'4 stops':4})
```

```python
final_df.head()
```

```python
#Now we are dropping the route column as we don't require two two informations, while the route may have some kind of
#effect on the price of the flight  but it is very complex to handle

final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
#Now we will handle duration and convert it all into minutes

final_df['Duration_hour']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
final_df['Duration_minutes']=final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
final_df.head()
```

```python
#Rplacing the NaN values in duration minutes with 0 as it is not actually NaN

final_df["Duration_minutes"]=final_df['Duration_minutes'].fillna('0',axis=0)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
#Somewhere there is 5m in hours and we need to handle it

final_df[final_df['Duration_hour']=='5m']
```

```python
#We can see that for two stops, the duration is 5 minutes lol, we will drop this

final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
#Now we see that for 5m the records are, now we will change the type again
```

```python
#Now we need to convert this into integer type

final_df['Duration_hour']=final_df["Duration_hour"].astype(int)
final_df['Duration_minutes']=final_df["Duration_minutes"].astype(int)
```

```python
final_df.info()
```

```python
#Now Duration can be dropped

final_df.drop('Duration',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Airline'].unique()
```

```python
final_df.info()
```

```python
#Now only categorical features are left and they can be label encoded

from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
final_df['Airline']=labelencoder.fit_transform(final_df['Airline'])
final_df['Source']=labelencoder.fit_transform(final_df['Source'])
final_df['Destination']=labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']=labelencoder.fit_transform(final_df['Additional_Info'])
```

```python
final_df.head()
```

```python
#This technique does not alas  capture the importance of the labels so we will one hot encode instead

pd.get_dummies(final_df,columns=['Airline','Source','Destination','Additional_Info'],drop_first=True)
```
