```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df_country.columns
```

```python
## Merging both the data frames
df_final=pd.merge(df,df_country,on='Country Code',how='left') #left means focusing on the left dataFrame

df_final.head()
```

```python
df_final.dtypes
```

```python
df_final.Country.value_counts()
```

```python
country_names=df_final.Country.value_counts().index
print(country_names)
```

```python
country_val=df_final.Country.value_counts().values
```

```python
##Pie chart

plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%') #Only top 3 countries 
```

Observations: The hioghest number of orders are placed from India on Zomato. Followed by USA and UK

```python
df_final.columns
```

```python
df_final.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings=df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Observations:
1. When rating is between 4.5 to 4.9--> Excellent
2. When ratings are between 4.0 to 4.4-->Very Good
3. When ratings are between 3.5 to 3.9-->Good
4. When ratings are between 2.5 to 3.4-->Average
5. When ratings are between 1.8 to 2.4-->Poor

Maximum People have not given a rating

```python
import matplotlib

matplotlib.rcParams['figure.figsize']=(12,6)  #Thius is just to increase the size of the diagrams

sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color",data=ratings)
#Note that these colors are not mapped
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color",data=ratings,palette=['white','red','orange','yellow','green',])
```

Observations:

1. Not rated count is the highest
2. Maximum number of ratings are between 2.5 to 3.4

```python
## Count Plot

sns.countplot(x="Rating color",data=ratings,palette=['white','red','orange','yellow','green','green'])

#This is just showing how frequently a color iws present in the dataframe
```
